```python
def vis_one_line(df):
    
    plt.figure(figsize=(11, 8))
    plt.xticks(np.arange(-1, 21, 1))
    plt.xticks(np.arange(-1, 21, 1))
    plt.title(df[1])
    sns.lineplot(data=df[0], x="cluster_num", y='before_PLL_s', label='before_PLL_s')
    sns.lineplot(data=df[0], x="cluster_num", y='one_sh', label='one_sh')
    #sns.lineplot(data=df, x="time", y="one_sh", label="one_sh")
    #sns.lineplot(data=df, x="cluster_num", y="one_osa", label="one_osa")
    #sns.lineplot(data=df, x="time", y="multi_sh", label="multi_sh")
    plt.show()
```

```python
data = [comp, rec, sci, misc, pol, rel]
name = ["comp", "rec", "sci", "misc", "pol", "rel"]
```

```python
for i in zip(data,name):
    vis_one_line(i)
```

```python
def vis_multi_line(df):
    plt.figure(figsize=(11, 8))
    plt.xticks(np.arange(-1, 21, 1))
    plt.xticks(np.arange(-1, 21, 1))
    sns.lineplot(data=df, x="cluster_num", y="one_sh", label="one_sh")
#     sns.lineplot(data=df, x="time", y="multi_lw", label="multi_lw")
    sns.lineplot(data=df, x="cluster_num", y="multi_sh", label="multi_sh")
#     sns.lineplot(data=df, x="time", y="multi_osa", label="multi_osa")
    plt.show()
```

```python
for i in [comp, rec, sci, misc, pol, rel]:
    vis_multi_line(i)
```

```python
def vis_silhouette_line(df):
    plt.figure(figsize=(11, 8))
    plt.xticks(np.arange(-1, 21, 1))
    plt.xticks(np.arange(-1, 21, 1))
    sns.lineplot(data=df, x="cluster_num", y="silhouette_a", label="silhouette_a")
#     sns.lineplot(data=df, x="time", y="multi_lw", label="multi_lw")
    sns.lineplot(data=df, x="cluster_num", y="silhouette_b", label="silhouette_b")
#     sns.lineplot(data=df, x="time", y="multi_osa", label="multi_osa")
    plt.show()
```

```python
for i in [comp, rec, sci, misc, pol, rel]:
    vis_silhouette_line(i)
```
