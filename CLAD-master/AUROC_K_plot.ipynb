```python
df = pd.read_csv("clinc_new.csv")
```

```python
df
```

```python
s = df[(df['method'] == "'SCCL'") & (df['LM'] == "'bert-base-uncased'") & (df['epoch'] == 1) 
       & (df['method_c'] == "'DEC_bert'") ]
```

```python
s.sort_values("cluster")
```

```python
plt.figure(figsize=(25, 15))
plt.rcParams['axes.titlepad'] = 50
graph = sns.lineplot(data=s, x='cluster', y ='AUROC',linewidth=7, dashes=False, label = "+ PLL")
graph.axhline(0.874, linewidth =3, c = "red",label = "Cluster + CL(BEST)")
graph.axhline(0.97, linewidth =3, c = "green",label = "Supervised")
graph.legend()
plt.legend(fontsize=40)
plt.title("Sensitivity of K", fontsize = 70)
plt.xlim((1,21))
plt.xticks(np.arange(10,210,10), fontsize = 50, rotation = 45)
plt.yticks(fontsize = 50)
plt.ylabel("AUROC",fontsize = 70)
plt.xlabel("Cluster #k",fontsize = 70)
```
